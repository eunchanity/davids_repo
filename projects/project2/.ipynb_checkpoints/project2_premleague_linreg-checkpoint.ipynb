# Linear Regression on Manchester United Players' Stats

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
import statsmodels.formula.api as smf
import patsy

import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import RidgeCV
%matplotlib inline
```

## 2018-2019 Stats

```python
premleague_df = pd.read_csv('project2_premleague_stats.csv', index_col='Unnamed: 0')
```

## Visualizing Data

```python
premleague_df.corr()
```

```python
sns.heatmap(premleague_df.corr(), cmap='seismic', annot=True, vmin=-1, vmax=1);
```

```python
sns.pairplot(premleague_df, height=1.2, aspect=1.5);
```

```python
# OLS with statsmodels
y, X = patsy.dmatrices('Gls ~ Age + Starts + Ast + CrdY + Shot + TotPassCmpPerc', data=premleague_df, return_type="dataframe")
model = sm.OLS(y, X)
fit = model.fit()
fit.summary()
```
